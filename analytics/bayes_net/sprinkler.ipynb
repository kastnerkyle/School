BAYESIAN NETWORKS WITH PYMC
===========================

An introductory example doing Bayesian networks in Python, using the PyMC framework.

```python
import pymc as mc
import numpy as np
```

This simple example shows how to use PyMC to simulate the "sprinkler problem". The goal is to determine if a sprinkler made the grass wet, or the rain, using the conditional probabilities below. This is a very simple problem, which is first step into Bayesian networks ("Hello World" for Bayes nets).

```python
from IPython.display import Image
Image(filename='bayesnet.png')
```

```python
G_obs = [1.]
N = len(G_obs)
 
R = mc.Bernoulli('R', .2, value=np.ones(N))
 
p_S = mc.Lambda('p_S', 
                lambda R=R: np.where(R, .01, .4),
                doc='Pr[S|R]')
S = mc.Bernoulli('S', p_S, value=np.ones(N))
 
p_G = mc.Lambda('p_G', 
                lambda S=S,
                R=R:np.where(S, np.where(R, .99, .9),
                             np.where(R, .8, 0.)),
                doc='Pr[G|S,R]')
G = mc.Bernoulli('G', p_G, value=G_obs, observed=True)
```

```python
m = mc.MCMC([R, p_S, S, p_G, G])
m.sample(iter=100000, burn=10000, thin=250)
```

```python
mc.Matplot.plot(m)
```

```python
mc.Matplot.summary_plot(m)
```

Judging by the above plots, it appears that the sprinkler is more likely to have caused the wet grass, although there is not enough information to generate confidence intervals for this example.
